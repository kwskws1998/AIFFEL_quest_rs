```python
from konlpy.tag import Mecab
import numpy as np
from collections import Counter

tokenizer = Mecab()
stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']

def load_data(train_data, test_data, num_words=10000):
    train_data.drop_duplicates(subset=['document'], inplace=True)
    train_data = train_data.dropna(how = 'any')
    test_data.drop_duplicates(subset=['document'], inplace=True)
    test_data = test_data.dropna(how = 'any')

    X_train = []
    for sentence in train_data['document']:
        temp_X = tokenizer.morphs(sentence) # 토큰화
        temp_X = [word for word in temp_X if not word in stopwords] # 불용어 제거
        X_train.append(temp_X)

    X_test = []
    for sentence in test_data['document']:
        temp_X = tokenizer.morphs(sentence) # 토큰화
        temp_X = [word for word in temp_X if not word in stopwords] # 불용어 제거
        X_test.append(temp_X)

    words = np.concatenate(X_train).tolist()
    counter = Counter(words)
    counter = counter.most_common(10000-4)
    vocab = ['', '', '', ''] + [key for key, _ in counter]
    word_to_index = {word:index for index, word in enumerate(vocab)}

    def wordlist_to_indexlist(wordlist):
        return [word_to_index[word] if word in word_to_index else word_to_index[''] for word in wordlist]

    X_train = list(map(wordlist_to_indexlist, X_train))
    X_test = list(map(wordlist_to_indexlist, X_test))

    return X_train, np.array(list(train_data['label'])), X_test, np.array(list(test_data['label'])), word_to_index

X_train, y_train, X_test, y_test, word_to_index = load_data(train_data, test_data)
```

```python
# 문장 1개를 활용할 딕셔너리와 함께 주면, 단어 인덱스 리스트 벡터로 변환해 주는 함수입니다.
# 단, 모든 문장은 <BOS>로 시작하는 것으로 합니다.
def get_encoded_sentence(sentence, word_to_index):
    return [word_to_index['<BOS>']]+[word_to_index[word] if word in word_to_index else word_to_index['<UNK>'] for word in sentence.split()]

# 여러 개의 문장 리스트를 한꺼번에 단어 인덱스 리스트 벡터로 encode해 주는 함수입니다.
def get_encoded_sentences(sentences, word_to_index):
    return [get_encoded_sentence(sentence, word_to_index) for sentence in sentences]

# 숫자 벡터로 encode된 문장을 원래대로 decode하는 함수입니다.
def get_decoded_sentence(encoded_sentence, index_to_word):
    return ' '.join(index_to_word[index] if index in index_to_word else '<UNK>' for index in encoded_sentence[1:])  #[1:]를 통해 <BOS>를 제외

# 여러 개의 숫자 벡터로 encode된 문장을 한꺼번에 원래대로 decode하는 함수입니다.
def get_decoded_sentences(encoded_sentences, index_to_word):
    return [get_decoded_sentence(encoded_sentence, index_to_word) for encoded_sentence in encoded_sentences]
```

```python
total_data_text = list(X_train) + list(X_test)
# 텍스트데이터 문장길이의 리스트를 생성한 후
num_tokens = [len(tokens) for tokens in total_data_text]
num_tokens = np.array(num_tokens)
# 문장길이의 평균값, 최대값, 표준편차를 계산해 본다.
print('문장길이 평균 : ', np.mean(num_tokens))
print('문장길이 최대 : ', np.max(num_tokens))
print('문장길이 표준편차 : ', np.std(num_tokens))

# 예를들어, 최대 길이를 (평균 + 2*표준편차)로 한다면,
max_tokens = np.mean(num_tokens) + 2 * np.std(num_tokens)
maxlen = int(max_tokens)
print('pad_sequences maxlen : ', maxlen)
print(f'전체 문장의 {np.sum(num_tokens < max_tokens) / len(num_tokens)}%가 maxlen 설정값 이내에 포함됩니다. ')

def pad_sequences(data, maxlen):
    padded_data = []
    for sentence in data:
        if len(sentence) < maxlen:
            # 0으로 패딩합니다.
            sentence = sentence + [0] * (maxlen - len(sentence))
        else:
            sentence = sentence[:maxlen]
        padded_data.append(sentence)
    return np.array(padded_data)

x_train_padded = pad_sequences(X_train, maxlen)
x_test_padded = pad_sequences(X_test, maxlen)

# PyTorch 모델에 입력하기 위한 텐서로 변환
x_train_tensor = torch.tensor(x_train_padded, dtype=torch.long)
x_test_tensor = torch.tensor(x_test_padded, dtype=torch.long)

print(x_train_tensor.shape)
```

Model 1-3 훈련 및 시각화 코드

```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import time

# --- 디바이스 설정 ---
device = torch.device("mps" if torch.mps.is_available() else "cpu")
print(f"Using device: {device}")

# --- 하이퍼파라미터 설정 ---
# 61599ba5 셀에서 생성된 word_to_index 사용
VOCAB_SIZE = len(word_to_index)
EMBEDDING_DIM = 128  # 단어 임베딩 벡터의 차원 (하이퍼파라미터)
LSTM_HIDDEN_DIM = 16   # LSTM의 은닉 상태 차원 (모델 1)
CNN_OUT_CHANNELS = 16 # CNN의 출력 채널 수 (모델 2)
OUTPUT_DIM = 1       # 최종 출력 차원 (긍정/부정)
NUM_EPOCHS = 7      # 학습 에포크 수 (시간 관계상 5로 설정)
LEARNING_RATE = 1e-3 # 학습률

print(f"Vocabulary Size: {VOCAB_SIZE}")
print(f"Embedding Dimension: {EMBEDDING_DIM}")

# --- 모델 1: LSTM ---
# (vocab_size와 word_vector_dim만 실제 값으로 변경)
class Model_1_LSTM(nn.Module):
    def __init__(self, vocab_size, word_vector_dim, hidden_dim, output_dim):
        super(Model_1_LSTM, self).__init__()
        self.embedding = nn.Embedding(vocab_size, word_vector_dim, padding_idx=PAD_INDEX)
        # batch_first=True : (batch_size, seq_len, feature_dim)
        self.lstm = nn.LSTM(word_vector_dim, hidden_dim, batch_first=True)
        self.fc1 = nn.Linear(hidden_dim, 8)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(8, output_dim)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        # x: (batch_size, seq_len)
        x = self.embedding(x)
        # x: (batch_size, seq_len, embedding_dim)
        
        # LSTM은 마지막 시퀀스의 (은닉 상태, 셀 상태)를 반환
        # lstm_out: (batch_size, seq_len, hidden_dim)
        # h_n (hidden_state): (num_layers, batch_size, hidden_dim)
        # c_n (cell_state): (num_layers, batch_size, hidden_dim)
        lstm_out, (h_n, c_n) = self.lstm(x)
        
        # 마지막 은닉 상태를 사용 (h_n[-1] == h_n[0] (단일 레이어이므로))
        x = h_n[-1] # (batch_size, hidden_dim)
        
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.sigmoid(x)
        return x.squeeze() # (batch_size,)

# --- 모델 2: 1D CNN ---
class Model_2_CNN(nn.Module):
    def __init__(self, vocab_size, word_vector_dim, out_channels, output_dim):
        super(Model_2_CNN, self).__init__()
        self.embedding = nn.Embedding(vocab_size, word_vector_dim, padding_idx=PAD_INDEX)
        
        # Conv1d는 (batch_size, in_channels, seq_len)을 기대
        # embedding_dim이 in_channels가 됨
        self.conv1 = nn.Conv1d(word_vector_dim, out_channels, kernel_size=7)
        self.conv2 = nn.Conv1d(out_channels, out_channels, kernel_size=7)
        self.pool = nn.MaxPool1d(5)
        self.global_max_pooling = nn.AdaptiveMaxPool1d(1) # 시퀀스 길이에 상관없이 가장 큰 값 1개만 선택
        
        self.fc1 = nn.Linear(out_channels, 8)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(8, output_dim)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        # x: (batch_size, seq_len)
        x = self.embedding(x)
        # x: (batch_size, seq_len, embedding_dim)
        
        # Conv1d를 위해 (batch_size, embedding_dim, seq_len)으로 변경
        x = x.permute(0, 2, 1)
        
        x = self.conv1(x) # (batch_size, out_channels, seq_len')
        x = self.relu(x)
        x = self.pool(x)
        
        x = self.conv2(x) # (batch_size, out_channels, seq_len'')
        x = self.relu(x)
        
        x = self.global_max_pooling(x) # (batch_size, out_channels, 1)
        
        # FC 레이어를 위해 flatten
        x = x.view(x.size(0), -1) # (batch_size, out_channels)
        
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.sigmoid(x)
        return x.squeeze() # (batch_size,)

# --- 모델 3: Global Max Pooling ---
class Model_3_GlobalPool(nn.Module):
    def __init__(self, vocab_size, word_vector_dim, output_dim):
        super(Model_3_GlobalPool, self).__init__()
        self.embedding = nn.Embedding(vocab_size, word_vector_dim, padding_idx=PAD_INDEX)
        self.global_max_pooling = nn.AdaptiveMaxPool1d(1)
        self.fc1 = nn.Linear(word_vector_dim, 8)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(8, output_dim)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        # x: (batch_size, seq_len)
        x = self.embedding(x)
        # x: (batch_size, seq_len, embedding_dim)
        
        # (batch_size, embedding_dim, seq_len)으로 변경
        x = x.permute(0, 2, 1)
        
        x = self.global_max_pooling(x) # (batch_size, embedding_dim, 1)
        
        # (batch_size, embedding_dim)으로 flatten
        x = x.view(x.size(0), -1) 
        
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.sigmoid(x)
        return x.squeeze() # (batch_size,)
```

```python
# 정확도 계산 함수
def calculate_accuracy(preds, y):
    # 0.5를 기준으로 0 또는 1로 변환
    rounded_preds = torch.round(preds)
    correct = (rounded_preds == y).float() # 정답이면 1, 아니면 0
    acc = correct.sum() / len(correct)
    return acc

# 모델 훈련 및 검증 함수
def train_model(model, train_loader, val_loader, optimizer, criterion, num_epochs, device):
    
    # 기록을 위한 딕셔너리
    history = {
        'train_loss': [],
        'train_acc': [],
        'val_loss': [],
        'val_acc': []
    }
    
    print(f"--- Training {model.__class__.__name__} ---")
    
    for epoch in range(num_epochs):
        start_time = time.time()
        
        # --- Training ---
        model.train() # 훈련 모드
        epoch_loss = 0
        epoch_acc = 0
        
        for inputs, labels in train_loader:
            inputs = inputs.to(device)
            labels = labels.to(device)
            
            # 1. 그래디언트 초기화
            optimizer.zero_grad()
            
            # 2. 순전파
            outputs = model(inputs)
            
            # 3. 손실 계산
            loss = criterion(outputs, labels)
            
            # 4. 정확도 계산
            acc = calculate_accuracy(outputs, labels)
            
            # 5. 역전파
            loss.backward()
            
            # 6. 가중치 업데이트
            optimizer.step()
            
            epoch_loss += loss.item()
            epoch_acc += acc.item()
            
        train_loss = epoch_loss / len(train_loader)
        train_acc = epoch_acc / len(train_loader)
        
        # --- Validation ---
        model.eval() # 평가 모드
        epoch_val_loss = 0
        epoch_val_acc = 0
        
        with torch.no_grad(): # 그래디언트 계산 비활성화
            for inputs, labels in val_loader:
                inputs = inputs.to(device)
                labels = labels.to(device)
                
                outputs = model(inputs)
                val_loss = criterion(outputs, labels)
                val_acc = calculate_accuracy(outputs, labels)
                
                epoch_val_loss += val_loss.item()
                epoch_val_acc += val_acc.item()
                
        val_loss = epoch_val_loss / len(val_loader)
        val_acc = epoch_val_acc / len(val_loader)
        
        end_time = time.time()
        
        # 기록
        history['train_loss'].append(train_loss)
        history['train_acc'].append(train_acc)
        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_acc)
        
        print(f"Epoch {epoch+1:02}/{num_epochs} | Time: {end_time-start_time:.2f}s")
        print(f"\tTrain Loss: {train_loss:.4f} | Train Acc: {train_acc*100:.2f}%")
        print(f"\tVal. Loss: {val_loss:.4f} | Val. Acc: {val_acc*100:.2f}%")
        
    return history

# 손실 함수 정의 (이진 분류)
criterion = nn.BCELoss()
```

```python
# 시각화 함수
def plot_history(history, model_name):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Loss 그래프
    ax1.plot(history['train_loss'], label='Train Loss')
    ax1.plot(history['val_loss'], label='Validation Loss')
    ax1.set_title(f'{model_name} - Loss')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    ax1.legend()
    
    # Accuracy 그래프
    ax2.plot(history['train_acc'], label='Train Accuracy')
    ax2.plot(history['val_acc'], label='Validation Accuracy')
    ax2.set_title(f'{model_name} - Accuracy')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Accuracy')
    ax2.legend()
    
    plt.suptitle(f"Model: {model_name}", fontsize=16)
    plt.show()

# --- 3개 모델 훈련 및 시각화 실행 ---
results = {} # 전체 결과를 저장할 딕셔너리

# 모델 1 (LSTM) 훈련
model_1 = Model_1_LSTM(VOCAB_SIZE, EMBEDDING_DIM, LSTM_HIDDEN_DIM, OUTPUT_DIM).to(device)
optimizer_1 = optim.Adam(model_1.parameters(), lr=LEARNING_RATE)
history_1 = train_model(model_1, train_loader, val_loader, optimizer_1, criterion, NUM_EPOCHS, device)
results['LSTM'] = history_1

# 모델 2 (CNN) 훈련
model_2 = Model_2_CNN(VOCAB_SIZE, EMBEDDING_DIM, CNN_OUT_CHANNELS, OUTPUT_DIM).to(device)
optimizer_2 = optim.Adam(model_2.parameters(), lr=LEARNING_RATE)
history_2 = train_model(model_2, train_loader, val_loader, optimizer_2, criterion, NUM_EPOCHS, device)
results['1D CNN'] = history_2

# 모델 3 (GlobalPool) 훈련
model_3 = Model_3_GlobalPool(VOCAB_SIZE, EMBEDDING_DIM, OUTPUT_DIM).to(device)
optimizer_3 = optim.Adam(model_3.parameters(), lr=LEARNING_RATE)
history_3 = train_model(model_3, train_loader, val_loader, optimizer_3, criterion, NUM_EPOCHS, device)
results['GlobalPool'] = history_3

# --- 결과 시각화 ---
print("\n" + "="*30)
print("     훈련 결과 시각화")
print("="*30)

for model_name, history in results.items():
    plot_history(history, model_name)
```

```python
# 필요한 라이브러리 import (이미 되어있다면 생략 가능)
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import os # os.getenv를 사용하기 위해 추가
from gensim.models import KeyedVectors # KeyedVectors를 사용하기 위해 추가

# --- 사전 훈련된 Gensim Word2Vec 모델 로드 ---
# 'word_vectors' 객체는 이 코드 실행 전에 이미 로드되어 있어야 합니다.
# 예시 로드 코드 (필요시 주석 해제)
word2vec_file_path = os.getenv("HOME") + '/Desktop/work/sentiment_classification/data/word2vec_ko.model'
try:
    word_vectors = KeyedVectors.load(word2vec_file_path)
    print("Gensim 모델 로딩 완료.")
except FileNotFoundError:
    print(f"오류: '{word2vec_file_path}'에서 모델을 찾을 수 없습니다.")
    raise
except NameError:
    print("오류: 'KeyedVectors'가 정의되지 않았습니다. 'from gensim.models import KeyedVectors'를 확인하세요.")
    raise
except Exception as e:
    print(f"모델 로드 중 오류 발생: {e}")
    raise


# --- word_vectors에서 어휘 목록과 벡터 행렬 생성 (한 번만 실행) ---
try:
    # Check if these exist from previous runs to avoid re-computation
    _ = index_to_word_pretrained
    _ = word_to_index_pretrained
    _ = all_vectors_pretrained
    print("사전 훈련된 모델의 어휘 매핑 및 벡터 행렬이 이미 존재합니다.")
except NameError:
    print("사전 훈련된 모델의 어휘 매핑 및 벡터 행렬 생성 중...")
    try:
        # Get the vocabulary list from the KeyedVectors object
        vocab_pretrained = list(word_vectors.wv.index_to_key)
        # Create mapping dictionaries
        index_to_word_pretrained = {i: word for i, word in enumerate(vocab_pretrained)}
        word_to_index_pretrained = {word: i for i, word in enumerate(vocab_pretrained)}
        # Get all vectors as a numpy matrix
        all_vectors_pretrained = word_vectors.wv.vectors
        print("어휘 매핑 및 벡터 행렬 생성 완료.")
    except NameError:
        print("\n오류: 'word_vectors' 객체가 없습니다. Word2Vec 모델을 먼저 로드해주세요.")
        # Stop execution here if word_vectors doesn't exist
        raise
    except Exception as e:
        print(f"\n어휘 매핑 생성 중 오류 발생: {e}")
        raise


def find_similar_words_explicit_cosine_pretrained(word, top_n=5):
    """
    사전 훈련된 Word2Vec 모델('word_vectors')을 사용하여
    sklearn.metrics.pairwise.cosine_similarity로 명시적으로
    코사인 유사도를 계산하여 유사 단어를 찾습니다.
    """
    try:
        if word not in word_to_index_pretrained:
            print(f"'{word}'는 사전 훈련된 Word2Vec 모델의 어휘 사전에 없습니다.")
            return

        word_idx = word_to_index_pretrained[word]
        # Get the vector for the input word from the pre-calculated matrix
        word_vec = all_vectors_pretrained[word_idx]

        # *** 명시적인 코사인 유사도 계산 ***
        sims = cosine_similarity([word_vec], all_vectors_pretrained)[0]

        # 자기 자신 제외, 상위 N개 인덱스 찾기
        # argsort() gives indices from lowest to highest similarity
        # We take the second highest up to top_n+1 highest (excluding self)
        top_indices = sims.argsort()[-(top_n + 1):-1][::-1] # Exclude self, reverse order

        print(f"\n--- '{word}'와(과) 유사한 단어 (사전 훈련된 Word2Vec, 직접 코사인 계산) ---")
        for idx in top_indices:
            # Check index bounds before accessing index_to_word_pretrained
            if idx < len(index_to_word_pretrained):
                 print(f"{index_to_word_pretrained[idx]} (유사도: {sims[idx]:.4f})")
            else:
                 print(f"Warning: Index {idx} out of bounds for index_to_word_pretrained")


    except NameError:
        print("오류: 'word_to_index_pretrained', 'index_to_word_pretrained', 또는 'all_vectors_pretrained'가 없습니다.")
        print("어휘 매핑 및 벡터 행렬 생성 부분을 먼저 실행해주세요.")
    except Exception as e:
        print(f"유사 단어 검색 중 오류 발생: {e}")

# --- 함수 실행 ---
# '영화' 단어와 유사한 단어 찾기
find_similar_words_explicit_cosine_pretrained('영화')

# '재미' 단어와 유사한 단어 찾기
find_similar_words_explicit_cosine_pretrained('재미')
```

```python
import os
import numpy as np
from gensim.models import KeyedVectors
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt # Make sure matplotlib is imported
import time # Make sure time is imported

# --- 1. Load Pre-trained Word2Vec Model & Create Weight Matrix ---
# Ensure word_vectors, word_to_index, and PAD_INDEX are loaded/defined
# from previous cells in your notebook.
# Example loading if needed:
word2vec_file_path = os.getenv("HOME") + '/Desktop/work/sentiment_classification/data/word2vec_ko.model'
print("Gensim Word2Vec 모델 로딩 중...")
try:
    word_vectors = KeyedVectors.load(word2vec_file_path)
    print("Gensim 모델 로딩 완료.")
except FileNotFoundError:
    print(f"오류: '{word2vec_file_path}'에서 모델을 찾을 수 없습니다.")
    raise
except Exception as e:
    print(f"모델 로딩 중 오류 발생: {e}")
    raise

try:
    VOCAB_SIZE = len(word_to_index)
    _ = PAD_INDEX # Check if PAD_INDEX exists
    _ = word_vectors # Check if word_vectors exists
except NameError:
    print("Error: 'word_to_index', 'PAD_INDEX', or 'word_vectors' not defined. Run previous cells.")
    raise

PRETRAINED_EMBEDDING_DIM = word_vectors.vector_size
print(f"사전 훈련된 Word2Vec 임베딩 차원: {PRETRAINED_EMBEDDING_DIM}")
print(f"Using Vocabulary Size: {VOCAB_SIZE}")

# Create weight matrix
weights_matrix = np.random.normal(scale=0.6, size=(VOCAB_SIZE, PRETRAINED_EMBEDDING_DIM))
weights_matrix = weights_matrix.astype(np.float32)
weights_matrix[PAD_INDEX] = np.zeros(PRETRAINED_EMBEDDING_DIM)

num_found = 0
for word, idx in word_to_index.items():
    if idx == PAD_INDEX: continue
    if idx >= weights_matrix.shape[0]:
        print(f"ERROR: Index {idx} for word '{word}' out of bounds ({weights_matrix.shape[0]}).")
        break
    try:
        weights_matrix[idx] = word_vectors.wv[word]
        num_found += 1
    except KeyError:
        pass
    except IndexError as e:
        print(f"IndexError for word '{word}' (idx {idx}): {e}")
        break

print(f"총 {VOCAB_SIZE}개 단어 중 {num_found}개의 벡터를 Word2Vec에서 로드했습니다.")

# Convert to PyTorch tensor
weights_tensor = torch.tensor(weights_matrix, dtype=torch.float32)

# --- 2. Ensure Models, Training/Plotting Functions, and Variables are Defined ---
# Make sure the classes Model_1_LSTM, Model_2_CNN, Model_3_GlobalPool,
# the functions train_model, plot_history, and variables like
# criterion, train_loader, val_loader, NUM_EPOCHS, LEARNING_RATE, device,
# LSTM_HIDDEN_DIM, CNN_OUT_CHANNELS, OUTPUT_DIM are available from previous cells.

# --- Accuracy Calculation Function ---
def calculate_accuracy(preds, y):
    rounded_preds = torch.round(preds)
    correct = (rounded_preds == y).float()
    acc = correct.sum() / len(correct)
    return acc

# --- Model Training and Validation Function ---
def train_model(model, train_loader, val_loader, optimizer, criterion, num_epochs, device):
    history = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}
    print(f"\n--- Training {model.__class__.__name__} with Pretrained Embeddings ---")
    for epoch in range(num_epochs):
        start_time = time.time()
        model.train()
        epoch_loss = 0
        epoch_acc = 0
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            acc = calculate_accuracy(outputs, labels)
            loss.backward()
            optimizer.step()
            epoch_loss += loss.item()
            epoch_acc += acc.item()
        train_loss = epoch_loss / len(train_loader)
        train_acc = epoch_acc / len(train_loader)

        model.eval()
        epoch_val_loss = 0
        epoch_val_acc = 0
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                val_loss = criterion(outputs, labels)
                val_acc = calculate_accuracy(outputs, labels)
                epoch_val_loss += val_loss.item()
                epoch_val_acc += val_acc.item()
        val_loss = epoch_val_loss / len(val_loader)
        val_acc = epoch_val_acc / len(val_loader)
        end_time = time.time()

        history['train_loss'].append(train_loss)
        history['train_acc'].append(train_acc)
        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_acc)

        print(f"Epoch {epoch+1:02}/{num_epochs} | Time: {end_time-start_time:.2f}s")
        print(f"\tTrain Loss: {train_loss:.4f} | Train Acc: {train_acc*100:.2f}%")
        print(f"\tVal. Loss: {val_loss:.4f} | Val. Acc: {val_acc*100:.2f}%")
    return history

# --- Visualization Function ---
def plot_history(history, model_name):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    ax1.plot(history['train_loss'], label='Train Loss')
    ax1.plot(history['val_loss'], label='Validation Loss')
    ax1.set_title(f'{model_name} - Loss')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    ax1.legend()
    ax2.plot(history['train_acc'], label='Train Accuracy')
    ax2.plot(history['val_acc'], label='Validation Accuracy')
    ax2.set_title(f'{model_name} - Accuracy')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Accuracy')
    ax2.legend()
    plt.suptitle(f"Model: {model_name}", fontsize=16)
    # Save the plot instead of showing it
    plot_filename = f"{model_name.replace(' ', '_').lower()}_pretrained_history.png"
    plt.savefig(plot_filename)
    print(f"Saved plot to {plot_filename}")
    plt.close(fig) # Close the figure to prevent display in output


# --- 3. Train and Plot All Models with Pre-trained Embeddings ---

results_pretrained = {} # Dictionary to store results

# --- Train Model 1 (LSTM) ---
try:
    model_1_pt = Model_1_LSTM(
        VOCAB_SIZE, PRETRAINED_EMBEDDING_DIM, LSTM_HIDDEN_DIM, OUTPUT_DIM
    ).to(device)
    model_1_pt.embedding.load_state_dict({'weight': weights_tensor})
    # model_1_pt.embedding.weight.requires_grad = False # Optional: Freeze
    optimizer_1_pt = optim.Adam(model_1_pt.parameters(), lr=LEARNING_RATE)
    history_1_pt = train_model(
        model_1_pt, train_loader, val_loader, optimizer_1_pt, criterion, NUM_EPOCHS, device
    )
    results_pretrained['LSTM_pretrained'] = history_1_pt
except NameError as e:
     print(f"Error during Model 1 training setup: {e}. Check variable/class definitions.")

# --- Train Model 2 (CNN) ---
try:
    model_2_pt = Model_2_CNN(
        VOCAB_SIZE, PRETRAINED_EMBEDDING_DIM, CNN_OUT_CHANNELS, OUTPUT_DIM
    ).to(device)
    model_2_pt.embedding.load_state_dict({'weight': weights_tensor})
    # model_2_pt.embedding.weight.requires_grad = False # Optional: Freeze
    optimizer_2_pt = optim.Adam(model_2_pt.parameters(), lr=LEARNING_RATE)
    history_2_pt = train_model(
        model_2_pt, train_loader, val_loader, optimizer_2_pt, criterion, NUM_EPOCHS, device
    )
    results_pretrained['CNN_pretrained'] = history_2_pt
except NameError as e:
    print(f"Error during Model 2 training setup: {e}. Check variable/class definitions.")

# --- Train Model 3 (GlobalPool) ---
try:
    model_3_pt = Model_3_GlobalPool(
        VOCAB_SIZE, PRETRAINED_EMBEDDING_DIM, OUTPUT_DIM
    ).to(device)
    model_3_pt.embedding.load_state_dict({'weight': weights_tensor})
    # model_3_pt.embedding.weight.requires_grad = False # Optional: Freeze
    optimizer_3_pt = optim.Adam(model_3_pt.parameters(), lr=LEARNING_RATE)
    history_3_pt = train_model(
        model_3_pt, train_loader, val_loader, optimizer_3_pt, criterion, NUM_EPOCHS, device
    )
    results_pretrained['GlobalPool_pretrained'] = history_3_pt
except NameError as e:
    print(f"Error during Model 3 training setup: {e}. Check variable/class definitions.")


# --- 4. Plot All Results ---
print("\n" + "="*30)
print("     훈련 결과 시각화 (Pretrained)")
print("="*30)

# Check if plot_history is defined before trying to plot
if 'plot_history' in locals() or 'plot_history' in globals():
    for model_name, history in results_pretrained.items():
        plot_history(history, model_name.replace('_pretrained', ' with Pretrained Word2Vec'))
else:
    print("Error: plot_history function is not defined. Cannot generate plots.")
```
